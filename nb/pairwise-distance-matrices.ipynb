# What do these images mean? 
First of all, each image above actually contains a single base image that's been repeated four times. (I just think the added symmetry makes things look nicer.) We'll be looking at this base image from now on. 

![](../images/pairwise_example_1.png)

To understand what the pixels mean, let's look at a small example. Imagine we break a song into one-second chunks, and measure the average sound amplitude in each chunk. (left figure)

Next, let's look at how chunks compare to each other by creating a grid of comparisons. We'll apply some comparison function, `f`, to each pair of chunks. (middle figure)


For our comparison function, let's use the absolute difference in amplitude:   $f(a,b) = \lvert a - b \rvert$

We can then color each cell according to the result of the comparison. In this case, cells with a larger absolute difference are darker. (right figure)






![](../images/pairwise_example.png)

Let's see how this looks when I create this graph using random data. 

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

First, let's generate some random data.

```python
np.random.seed(24124124)
chunk_amplitudes = np.random.random(50)

plt.plot(chunk_amplitudes,'-k')

plt.ylabel('amplitude')
plt.xlabel('chunk index')
plt.title('chunk amplitudes (randomly-generated)')
plt.tight_layout()
```

Now, let's create the grid of comparisons. 

```python
from sklearn.metrics.pairwise import pairwise_distances

absolute_difference = lambda a,b: np.abs(a - b)

comparison_grid = pairwise_distances(
    chunk_amplitudes.reshape(-1,1), # need to reshape 1-D array into 2-D array with one column
    metric = absolute_difference)

comparison_grid[:3,:3]
```

And now we can plot the grid of comparisons.

```python
plt.figure(figsize =(5,5))
plt.imshow(comparison_grid, cmap = 'Greys');
plt.colorbar()
plt.tight_layout()
```

For fun, we can create a version of this plot that's repeated four times, just like the original pictures. 

```python
def quad_repeat(matrix):
    top_left = matrix
    top_right = np.fliplr(matrix) # flip left-right
    bottom_left = np.flipud(matrix) # flip up-down
    bottom_right = np.flipud(np.fliplr(matrix))
    
    top_row = np.hstack([top_left, top_right]) # horizontal stack
    bottom_row = np.hstack([bottom_left, bottom_right]) # vertical stack
    
    return np.vstack([top_row, bottom_row])
```

```python
plt.figure(figsize =(5,5))
plt.imshow(quad_repeat(comparison_grid), cmap = 'Greys');
plt.colorbar()
plt.tight_layout()
```

## Putting it all together

```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
from sklearn.metrics.pairwise import pairwise_distances
```

### Random version

```python
x = np.random.random((100, 2))
matrix = pairwise_distances(x)
plt.imshow(matrix, cmap = 'Greys')
```

### Real music version

```python
# Load data with Librosa
y, sample_rate = librosa.load("../audio/zella.wav")
chroma = librosa.feature.chroma_cqt(y = y, sr = sample_rate) # ignore the warning

# Prepare data with Numpy
n_rows, n_columns = chroma.shape
bin_edges = np.linspace(0, n_columns, 100).astype(int)
binned_chroma = librosa.util.sync(chroma, bin_edges)

# Compute pairwise distances
matrix = pairwise_distances(binned_chroma.T)

# Plot matrix
plt.figure(figsize = (10,10))
plt.imshow(quad_repeat(matrix), cmap = 'Greys')
plt.xticks([])
plt.yticks([])
plt.axis('off');
```
